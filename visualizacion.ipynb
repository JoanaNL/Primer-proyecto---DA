```python
import pandas as pd
import os

carpeta = 'C:\\Users\\Haine\\Proyecto 1 abril\\Primer-proyecto---DA\\Dataset'

archivos_csv = [f for f in os.listdir(carpeta) if f.endswith((
    'df_mineria_datos_edades.csv',
    'df_mineria_datos_inmigracion.csv',
    'df_mineria_datos_vivienda.csv'
))]

dataframes = []

for archivo in archivos_csv:
    ruta = os.path.join(carpeta, archivo)
    try:
        df = pd.read_csv(ruta, on_bad_lines='skip', engine='python')
        df['archivo_origen'] = archivo  # agrega una columna para rastrear el origen
        dataframes.append(df)
        print(f"✅ Cargado: {archivo} ({df.shape[0]} filas)")
    except Exception as e:
        print(f"❌ Error al cargar {archivo}: {e}")

# Une todos los DataFrames exitosamente cargados
df_unido = pd.concat(dataframes, ignore_index=True)

print(f"\n✅ Unión completada. Total de filas: {df_unido.shape[0]}")
```
